```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display

# Define grading scheme
grade_to_numeric = {
    'A': 94, 'A-': 91,
    'B+': 88, 'B': 85
}

# Define weights
weights = {
    'Section': 25,
    'Quiz Choice': 6,
    'Quiz Essay': 14,
    'Policy': 20,
    'Simulation': 10,
    'Term Paper': 30
}

# Initialize dataframe with categories and weights
df = pd.DataFrame({
    'Category': list(weights.keys()),
    'Weight': list(weights.values()),
    'Grade': '',
    'Numeric Grade': '',
    'Weighted Score': 0
})

# Function to update numeric grade and weighted score based on input
def update_grades(change):
    for i, (index, row) in enumerate(df.iterrows()):
        grade_input = grade_inputs[i].value.upper()
        if grade_input in grade_to_numeric:
            df.at[index, 'Grade'] = grade_input
            df.at[index, 'Numeric Grade'] = grade_to_numeric[grade_input]
            df.at[index, 'Weighted Score'] = grade_to_numeric[grade_input] * (row['Weight'] / 100)
        else:
            df.at[index, 'Grade'] = ''
            df.at[index, 'Numeric Grade'] = ''
            df.at[index, 'Weighted Score'] = 0
    score = df['Weighted Score'].sum()
    final_grade_label.value = f"Final Grade: {get_final_grade(score)}"

# Function to determine the final grade based on total score
def get_final_grade(score):
    if 95 <= score <= 100:
        return 'A'
    elif 90 <= score <= 94:
        return 'A-'
    elif 86 <= score <= 89:
        return 'B+'
    elif 82 <= score <= 85:
        return 'B'
    else:
        return 'Below B'

# Create input widgets for grades
grade_inputs = [widgets.Combobox(options=['', 'A', 'A-', 'B+', 'B'],
                                 placeholder='Select Grade',
                                 ensure_option=True,
                                 layout={'width': 'max-content'}) for _ in weights]

for grade_input in grade_inputs:
    grade_input.observe(update_grades, names='value')

# Display inputs
inputs_display = widgets.VBox([widgets.HBox([widgets.Label(value=f"{category}: "), grade_input])
                               for category, grade_input in zip(weights.keys(), grade_inputs)])
display(inputs_display)

# Display final grade
final_grade_label = widgets.Label(value="Final Grade: ")
display(final_grade_label)
```
